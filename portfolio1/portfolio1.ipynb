## Task 4 [3 marks]
Using bowtie2 , featureCounts and your own processing of the results, identify the three genes with the highest number of mapped reads.
Apart from the input and output files, for this task you should use the default parameters of both tools.

```python
!bowtie2 -x e_coli -S output.sam reads_042_R1.fastq
!featureCounts -a e_coli.gtf -o fc_output output.sam
```

```python
result = {}

# Read fc_output line by line and count the number of reads that map to each gene
with open('fc_output', 'r') as f:
    # Start on line 3 
    for line in f.readlines()[2:]:
        if line:
            # Split the line by tabs and get the gene name
            gene = line.split('\t')[0]
            count = int(line.split('\t')[-1].strip())
            result[gene] = count

# Sort the result dictionary by value and print the top 10 genes
result = sorted(result.items(), key=lambda x: x[1], reverse=True)
result
```
